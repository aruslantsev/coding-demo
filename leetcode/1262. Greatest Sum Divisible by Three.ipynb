Given an integer array nums, return the maximum possible sum of elements of the array such that it is divisible by three.

 

Example 1:

Input: nums = [3,6,5,1,8]
Output: 18
Explanation: Pick numbers 3, 6, 1 and 8 their sum is 18 (maximum sum divisible by 3).

Example 2:

Input: nums = [4]
Output: 0
Explanation: Since 4 is not divisible by 3, do not pick any number.

Example 3:

Input: nums = [1,2,3,4,4]
Output: 12
Explanation: Pick numbers 1, 3, 4 and 4 their sum is 12 (maximum sum divisible by 3).

 

Constraints:

    1 <= nums.length <= 4 * 104
    1 <= nums[i] <= 104



```python
class Solution:
    def maxSumDivThree(self, nums: List[int]) -> int:
        nums.sort()
        ans = sum(nums)
    
        r1 = [num for num in nums if num % 3 == 1]
        r2 = [num for num in nums if num % 3 == 2]
        d11 = r1[0] if r1 else None
        d22 = r2[0] if r2 else None
        d21 = sum(r1[:2]) if len(r1) >= 2 else None
        d12 = sum(r2[:2]) if len(r2) >= 2 else None

        if d11 and d12:
            s1 = min(d11, d12)
        elif d11:
            s1 = d11
        elif d12:
            s1 = d12
        else:
            s1 = None

        if d21 and d22:
            s2 = min(d21, d22)
        elif d21:
            s2 = d21
        elif d22:
            s2 = d22
        else:
            s2 = None

        if ans % 3 == 0:
            return ans
        elif ans % 3 == 1:
            if s1:
                return ans - s1
            else:
                return 0
        else:
            if s2:
                return ans - s2
            else:
                return 0
```

```python
class Solution:
    def maxSumDivThree(self, nums: List[int]) -> int:
        nums.sort()
        ans = sum(nums)
    
        r1 = [num for num in nums if num % 3 == 1]
        r2 = [num for num in nums if num % 3 == 2]

        remove = float("inf")
        if ans % 3 == 0:
            remove = 0
        elif ans % 3 == 1:
            if r1:
                remove = min(remove, r1[0])
            if len(r2) >= 2:
                remove = min(remove, sum(r2[:2]))
        else:
            if len(r1) >= 2:
                remove = min(remove, sum(r1[:2]))
            if r2:
                remove = min(remove, r2[0])
        
        return max(ans - remove, 0)
```

```python
class Solution:
    def maxSumDivThree(self, nums: List[int]) -> int:
        f = [0, -float("inf"), -float("inf")]
        for num in nums:
            g = f[:]
            for i in range(3):
                rem = num % 3
                g[(i + rem) % 3] = max(f[i] + num, f[(i + rem) % 3])
            f = g
        return f[0]
```
