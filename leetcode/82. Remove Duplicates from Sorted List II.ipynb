Given the head of a sorted linked list, delete all nodes that have duplicate numbers, leaving only distinct numbers from the original list. Return the linked list sorted as well.

 

Example 1:



Input: head = [1,2,3,3,4,4,5]
Output: [1,2,5]

Example 2:



Input: head = [1,1,1,2,3]
Output: [2,3]

 

Constraints:

    The number of nodes in the list is in the range [0, 300].
    -100 <= Node.val <= 100
    The list is guaranteed to be sorted in ascending order.


```python
from typing import Optional

# Definition for singly-linked list.
class ListNode:
    def __init__(self, val=0, next=None):
        self.val = val
        self.next = next

class Solution:
    def deleteDuplicates(self, head: Optional[ListNode]) -> Optional[ListNode]:
        if head is None:
            return None
        
        ret = start = None
        bad_val = None
        
        while head is not None:
            if (head.next is not None) and (head.val == head.next.val):
                bad_val = head.val                
            elif head.val != bad_val:
                bad_val = None
                if start:
                    start.next = ListNode(head.val, None)
                    start = start.next
                else:
                    ret = start = ListNode(head.val)
                    
            head = head.next
        
        return ret
        
```

```python
s = Solution()
res = s.deleteDuplicates(
    ListNode(1, ListNode(2, ListNode(3, ListNode(3, ListNode(4, ListNode(4, ListNode(5))))))))
while res:
    print(res.val)
    res = res.next
```

```python
res = s.deleteDuplicates(
    ListNode(1, ListNode(1, ListNode(1, ListNode(2, ListNode(3))))))
while res:
    print(res.val)
    res = res.next
```
