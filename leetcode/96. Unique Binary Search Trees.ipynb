Given an integer n, return the number of structurally unique BST's (binary search trees) which has exactly n nodes of unique values from 1 to n.

 

Example 1:



Input: n = 3
Output: 5

Example 2:

Input: n = 1
Output: 1

 

Constraints:

    1 <= n <= 19



```python
class Solution:
    @cache
    def numTrees(self, n: int) -> int:
        if n <= 1:
            return 1
        ans = 0
        for i in range(n):
            left_trees = self.numTrees(i)
            right_trees = self.numTrees(n - 1 - i)
            ans += left_trees * right_trees
        return ans
```
