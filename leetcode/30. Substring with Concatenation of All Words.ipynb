You are given a string s and an array of strings words of the same length. Return all starting indices of substring(s) in s that is a concatenation of each word in words exactly once, in any order, and without any intervening characters.

You can return the answer in any order.

 

Example 1:

Input: s = "barfoothefoobarman", words = ["foo","bar"]
Output: [0,9]
Explanation: Substrings starting at index 0 and 9 are "barfoo" and "foobar" respectively.
The output order does not matter, returning [9,0] is fine too.

Example 2:

Input: s = "wordgoodgoodgoodbestword", words = ["word","good","best","word"]
Output: []

Example 3:

Input: s = "barfoofoobarthefoobarman", words = ["bar","foo","the"]
Output: [6,9,12]

 

Constraints:

    1 <= s.length <= 104
    1 <= words.length <= 5000
    1 <= words[i].length <= 30
    s and words[i] consist of lowercase English letters.



```python
from typing import List
from collections import Counter

class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        def compare_dicts(s, w):
            for k, v in w.items():
                if k not in s or v != s[k]:
                    return False
            return True
        
        words = ''.join(words)
        words_count = dict(Counter(words))
        if len(words) > len(s):
            return []
        
        s_count = dict(Counter(s[:len(words)]))
        for key in s_count.keys():
            if key not in words_count:
                del s_count[key]
        for key in words_count.keys():
            if key not in s_count:
                s_count[key] = 0
        ans = []
        
        if compare_dicts(s_count, words_count):
            ans.append(0)
        
        for i in range(len(words), len(s)):
            if s[i - len(words)] in words_count:
                s_count[s[i - len(words)]] -= 1
            if s[i] in words_count:
                s_count[s[i]] += 1
                
            if compare_dicts(s_count, words_count):
                ans.append(i)
                
        return ans
```

```python
from collections import Counter, defaultdict

class Solution:
    def findSubstring(self, s: str, words: list[str]) -> list[int]:
        m, n, wordLength = len(s), len(words), len(words[0])
        words = Counter(words)

        def counts(sub):
            count = defaultdict(int)
            for i in range(0, n * wordLength, wordLength):
                
                word = sub[i : i + wordLength]
                if word in words:
                    count[word] += 1
                else:
                    return {}

            return count

        res = []
        for i in range(0, m - n * wordLength + 1):
            if counts(s[i : i + n * wordLength]) == words:
                res.append(i)

        return res
```

```python
class Solution:
    def findSubstring(self, s: str, words: List[str]) -> List[int]:
        m = len(words)
        k = len(words[0])
        res = []
        
        for i in range(k):
            left = i
            d = Counter(words)
            
            for j in range(left, len(s) + 1 - k, k):
                word = s[j: j + k]
                d[word] -= 1
                
                while d[word] < 0:
                    d[s[left: left + k]] += 1
                    left += k
                if left + k * m == j + k:
                    res.append(left)
        
        return res
```
